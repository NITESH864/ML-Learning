```python
# Import essential libraries 
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Import dataset 
df=pd.read_csv("student_info.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
df=df.fillna(df.mean())
df.isnull().sum()
```

```python
#Visualization of data
plt.scatter(x=df.study_hours,y=df.student_marks,color='y')
plt.xlabel("Study Hours")
plt.ylabel("Student Matrks")
plt.title("Student Marks Predictors")
plt.grid(color='r')
plt.show()
```

```python
# separate independent and dependent variables

x=df.drop("student_marks",axis=1)
y=df.drop("study_hours",axis=1)
```

```python
x.head()
```

```python
y.head()
```

```python
# Train-Test Splitting

from sklearn.model_selection import train_test_split
x_train,x_test, y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#Apply Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
# Test the model 
res=lr.predict([[8.0]])[0][0].round(2)
print(res)
```

```python
# Creation of .pkl file as extract of model
import pickle
pickle.dump(lr,open("smp.pkl","wb"))
print("Model is saved")
```
