```python
#Import essential library
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
#Read csv file
df=pd.read_csv("iris.csv")
df.head()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Seprate dataset into dependent and independent veriables
X=df.iloc[:,:4].values
y=df.iloc[:,4].values
```

```python
# Train-test
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
# Feature scaling-
from sklearn.preprocessing import StandardScaler
ss=StandardScaler()
X_train=ss.fit_transform(X_train)
X_test=ss.fit_transform(X_test)
```

### Feature Scaling ko aur aasan bhasha mein samjho (Real-life Example)
Socho aapke paas ek race hai jisme kuch log meter mein daudte hain, kuch centimeter mein, aur kuch kilometer mein. Agar sabko ek hi unit mein convert nahi karoge, toh race ka result galat ho jayega!
Machine learning mein bhi aisa hota hai:
- Kuch features (columns) chhote number mein hote hain, kuch bade number mein.
- Agar model ko sab kuch ek hi scale par nahi milega, toh wo confuse ho jayega aur sahi prediction nahi karega.
StandardScaler kya karta hai?
- Har feature ko aise scale karta hai ki uska average (mean) 0 ho jaaye aur spread (standard deviation) 1 ho jaaye.
- Isse sabhi features ek hi level par aa jaate hain, aur model ko training mein aasani hoti hai.
Yahi reason hai ki feature scaling karna important hai, taaki model sabhi data ko barabar importance de sake!

```python
# Apply decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dtc=DecisionTreeClassifier()
dtc.fit(X_train,y_train)
```

```python
# Test the model
y_pred=dtc.predict(X_test)
y_pred
```

```python
# Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
cm
```
