```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')
```

```python
x=[1,2,3]
y=[4,5,1]
plt.plot(x,y,'r',linewidth=2)
plt.title('mygraph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True,color='k')
plt.show()
```

```python
# Scatter Plot
df=pd.read_csv("student_info.csv")
df.head()
```

```python
plt.scatter(x=df.study_hours,y=df.student_marks)
plt.title("student Marks Prediction")
plt.xlabel("study_Hours")
plt.ylabel("student_marks")
#plt.grid(True,color='k')
plt.show()
```

```python
import matplotlib.pyplot as plt
from matplotlib import style
x=[5,8,10]
y=[12,16,6]
x1=[6,9,11]
y1=[6,15,11]
plt.title('Student_info')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
style.use('ggplot')
plt.plot(x,y,'g',label='line one',linewidth=2)
plt.plot(x1,y1,'r',label='line two',linewidth=2)
plt.legend()
plt.grid(True,color='b')
plt.show()
```

```python
#Bar graph
days=[0.25,1.25,2.25,3.25,4.25]
dist=[50,40,70,80,20]
days2=[0.75,1.75,2.75,3.75,4.75]
dist2=[80,20,20,50,60]
plt.bar(days,dist,label='BMW',width=.5)
plt.bar(days2,dist2,label='AUDI',width=.5)
plt.title('Information about Cars')
plt.xlabel("Days")
plt.ylabel("Distance")
plt.legend()
plt.show()
```
